### Ridge regression and LASSO for one variable

In this problem weâ€™ll try to understand the effect of regularization on model parameters by
considering what happens when n=p=1, i.e. we have just one data point and just one predictor. In this
case the equations are simple and by plotting the results we can build intuition.

```python
# Part A
import numpy as np
import matplotlib.pyplot as plt

def C_rigid(y, beta, lamda):
    return (y-beta)**2 + lamda*beta**2;

betas = np.arange(0.0,3.0, 0.1)
plt.plot(betas, [ C_rigid(2,beta,0) for beta in betas ],color="red",linewidth=2)
plt.title("the cost function of Ridge Regression")
plt.xlabel("beta")
plt.ylabel("C_rigid")
plt.show()
```

#### What is beta, the value of  beta that minimizes Cridge=  (2-beta)^2 ?

Since the differentail of Cridge is 
d(Cridge)/d(beta) = -2(2-beta)

so this is minimized when beta is 2


```python
#part B 
# Now plot Cridge as a function of Î² for Î» = 1 

betas = np.arange(0.0,3.0, 0.1)
plt.plot(betas, [ C_rigid(2,beta,1) for beta in betas ],color="red",linewidth=2)
plt.title("the cost function of Ridge Regression")
plt.xlabel("beta")
plt.ylabel("C_rigid")
plt.show()
```

#### Show that the new beta is given by the formula in (6.14), . Describe, in words, how changes as Î²Ë† /(1 ) /2 Î²Ë† = y + Î» = y Î²Ë† Î» increases -- this illustrates why ridge regression is a form of shrinkage .

The new beta that min the cost function is beta=1


since now Cridge = (2- beta)^2 + gama * beta^2

d(Cridge)/d(beta) = -2(2-beta) + 2* beta^2

Now if gama ==1
then beta = y/2

As gama increase the beta will decrease and shrink to 0

```python
# part C
# Consider the cost function for LASSO, with p=1:
# Classo = (y âˆ’ Î²)2 + Î» |Î²|
# Make plots showing Classo as a function of Î² for Î» = 0 and Î» = 1 .


def C_lasso(y, beta, lamda):
    return (y-beta)**2 + lamda*abs(beta);

betas = np.arange(0.0,3.0, 0.1)
plt.plot(betas, [ C_lasso(2,beta,0) for beta in betas ],color="red",linewidth=2)
plt.plot(betas, [ C_lasso(2,beta,1) for beta in betas ],color="blue",linewidth=2)
plt.title("the cost function of lasso")
plt.xlabel("beta")
plt.ylabel("C_lasso")
plt.show()
```

The min value of beta for lamda=0 is  beta=2
when Lamda=0 it shows the cost func without regularization

The min value of beta for lamda=1 is  beta=1.5

Notice that the two cost functions converge at beta=0, which is a result of weight shrinkage.

```python
# Part D
# Fit a simple linear regression model of the form, BehaviorResponse ~ 1 + F0Hz_1. 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from patsy import dmatrices

### do linear regression
y, X = dmatrices('BehaviorResponse ~ 1 + F0Hz_1', data=df, return_type='dataframe')
X = sm.add_constant(X)
mod = sm.OLS(y, X)
# fit model
res = mod.fit()
# look at results
print(res.summary())
```

#### Is the slope parameter statistically significant?
Since the p-value is smaller than 0.05, there is a statistically significant relationship between BehaviorResponse and power at 0Hz

```python
# part F
# Fit a 103 models and find the single features with the lowest MSE
# for each fit use the MSE
# make a plot showing MSE VS feature number 

mse_list = [0 for i in range(103)]
feature_list = []
for feature in features:
    if feature != "Time" and feature != "BehaviorResponse":
        feature_list.append(feature)
        
# 103 iterations
index = 0
for feature in feature_list:
    y, X = dmatrices('BehaviorResponse ~' + feature, data=df, return_type='dataframe')
    X = sm.add_constant(X)
    mod = sm.OLS(y, X)
    # fit model
    res = mod.fit()
    #predict the data
    y_hat = res.predict(X);
    mse_list[index] = np.mean( (df['BehaviorResponse'] - y_hat)**2 )
    index += 1
```

```python
print min(mse_list)
print "So F11Hz_10 is winning feature"

x=range(103)
y= []
for index in range(103):
    y.append( mse_list[index] )

plt.plot(x,y)
plt.title("MSE  VS Feature Number")
plt.xlabel("Feature Number")
plt.ylabel("MSE")
plt.show()
```
